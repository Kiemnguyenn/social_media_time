```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# 1. Read data, display general information of data
df_socialMediaTime = pd.read_csv('data/dummy_data.csv')
df_socialMediaTime.head()
```

```python
df_socialMediaTime.tail()
```

```python
df_socialMediaTime.info()
```

```python
df_socialMediaTime.describe()
```

```python
# 2. Average time users spend on social networks across all data 
plt.figure(figsize = (8, 5))
sns.barplot(data = df_socialMediaTime, x= 'time_spent', y = df_socialMediaTime.index, estimator = 'count', errorbar = None)
plt.ylabel('Total Number of Users')
plt.show()
```

```python
avg_time_on_social_media = df_socialMediaTime.groupby(by = ['platform']).agg({'time_spent':'mean'}).reset_index()
avg_time_on_social_media
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(data = avg_time_on_social_media, x = 'platform', y = 'time_spent', errorbar = None)
plt.show()
```

```python
# 3. Analyze time spent using social networks by age 
# Age groups
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64']
df_socialMediaTime['age_group'] = pd.cut(df_socialMediaTime['age'], bins = bins, labels = labels, right = False)
# Calculate the average time spent using social networks for each age group
avg_time_spent_by_age_group = df_socialMediaTime.groupby('age_group', observed = False)['time_spent'].mean()
# Add an avg_time_spent_by_age_group column to the data
df_socialMediaTime['avg_time_spent_by_age_group'] = df_socialMediaTime['age_group'].map(avg_time_spent_by_age_group)
df_socialMediaTime.head()
```

```python
unique_values = df_socialMediaTime[['age_group', 'avg_time_spent_by_age_group']].drop_duplicates()
unique_values
```

```python
sorted_unique_values = unique_values.sort_values(by = 'age_group')
sorted_unique_values
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(sorted_unique_values['age_group'], unique_values['avg_time_spent_by_age_group'], color = 'r', marker = 'o')
plt.xlabel('Age Group')
plt.ylabel('Average Time Spent')
plt.title('Average Time Spent on Social Media by Age Group')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(sorted_unique_values['age_group'], unique_values['avg_time_spent_by_age_group'], marker = 'o', color = 'b', linestyle = '-')
plt.xlabel('Age Group')
plt.ylabel('Average Time Spent')
plt.title('Average Time Spent on Social Media by Age Group')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
custom_palette = {'Facebook': '#0000FF', 'Instagram': '#FF00FF', 'YouTube': '#FF0000'}
sns.countplot(data = df_socialMediaTime, x = 'age_group', hue = 'platform', palette = custom_palette)
plt.xlabel('Age Group')
plt.ylabel('Total Number of Users')
plt.title('Social Media Usage by Age Group')
plt.legend(title = 'Platform')
plt.show()
```

```python
# 4. Comparison of time spent using social networks by gender
avg_time_spent = df_socialMediaTime.groupby(['gender', 'platform'])['time_spent'].mean()
df_socialMediaTime['avg_time_spent'] = df_socialMediaTime.apply(lambda row: avg_time_spent[(row['gender'], row['platform'])], axis = 1)
df_socialMediaTime.head()
```

```python
unique_values_1 = df_socialMediaTime[['gender', 'avg_time_spent']].drop_duplicates()
unique_values_1
```

```python
plt.figure(figsize = (10, 7))
custom_palette = {'Facebook': '#0000FF', 'Instagram': '#FF00FF', 'YouTube': '#FF0000'}
sns.barplot(data = df_socialMediaTime, x = 'gender', y = 'avg_time_spent', hue = 'platform', palette = custom_palette, linewidth = 0)
plt.xlabel('Gender')
plt.ylabel('Average Time Spent')
plt.title('Time Spent Using Social Networks by Gender')
plt.legend(title = 'Platform', bbox_to_anchor = (1, 1),loc = 'upper left')
plt.show()
```

```python
# 5. Comparison of time spent using social networks by demographics 
avg_time_spent = df_socialMediaTime.groupby(['demographics', 'platform'])['time_spent'].mean()
df_socialMediaTime['avg_time_spent_by_demographics'] = df_socialMediaTime.apply(lambda row: avg_time_spent[(row['demographics'], 
                                                                                                            row['platform'])], axis = 1)
df_socialMediaTime.head()
```

```python
unique_values_2 = df_socialMediaTime[['demographics', 'avg_time_spent_by_demographics']].drop_duplicates()
unique_values_2
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(data = df_socialMediaTime, x = 'demographics', y = 'avg_time_spent_by_demographics')
plt.title('Average Time Spent on Social Media by Demographics')
plt.xlabel('Demographics')
plt.ylabel('Average Time Spent')
plt.show()
```

```python
# 6. The influence of different factors on social network usage habits
variables = ['age_group', 'time_spent', 'gender', 'interests', 'location', 'demographics', 'profession', 'income', 'indebt', 'isHomeOwner', 
             'Owns_Car', 'avg_time_spent']
platforms = ['facebook', 'youtube', 'instagram']
rows = 4
cols = 3
fig, axes = plt.subplots(rows, cols, figsize=(15, 15))
for i, variable in enumerate(variables):
    row = i // cols
    col = i % cols
    ax = axes[row, col]
    pivot_table = df_socialMediaTime.pivot_table(index = variable, columns = 'platform', aggfunc = 'size')
    sns.heatmap(pivot_table, cmap = 'Reds', annot = True, fmt = 'g', ax = ax)
    ax.set_title(f'{variable.capitalize()} Heatmap')
    ax.set_xlabel('Platform')
    ax.set_ylabel(variable.capitalize())
plt.tight_layout()
plt.show()
```

```python
# 7. Compare countries and identify trends
total_users_by_location = df_socialMediaTime.groupby('location').size().reset_index(name = 'Total Number of Users')
total_users_by_location
```

```python
variables = ['age_group', 'time_spent', 'gender', 'interests', 'platform', 'demographics', 'profession', 'income', 'indebt', 'isHomeOwner', 
             'Owns_Car', 'avg_time_spent']
countries = ['United Kingdom', 'United States', 'Australia']
rows = 4
cols = 3
fig, axes = plt.subplots(rows, cols, figsize = (15, 20))
for i, variable in enumerate(variables):
    row = i // cols
    col = i % cols
    ax = axes[row, col]
    pivot_table = df_socialMediaTime[df_socialMediaTime['location'].isin(countries)].pivot_table(index = variable, columns = 'location', aggfunc = 'size')
    sns.heatmap(pivot_table, cmap='Reds', annot=True, fmt='g', ax=ax)
    ax.set_title(f'{variable.capitalize()} Heatmap')
    ax.set_xlabel('Country')
    ax.set_ylabel(variable.capitalize())
plt.tight_layout()
plt.show()
```
